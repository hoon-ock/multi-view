```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Enhancement strategies and corresponding MAE values
strategies = ['Baseline', 'GAP', 'CA', 'CA+GAP']
mae_values = [0.713, 0.643, 0.378, 0.346]
mae_errors = [0.014, 0.020, 0.005, 0.005]

# Horizontal lines for DimeNet++, GemNet-OC, and EquiformerV2
horizontal_lines = {
    'DimeNet++': 0.326,
    'GemNet-OC': 0.177,
    'EquiformerV2': 0.132
}

# Creating the bar plot
plt.figure(figsize=(10, 6))
bar_positions = np.arange(len(strategies))
plt.bar(bar_positions, mae_values, yerr=mae_errors, capsize=5, color='skyblue')

# Adding horizontal lines
for model, value in horizontal_lines.items():
    plt.axhline(y=value, linestyle='--', color='grey', linewidth=1)
    plt.text(len(strategies), value, 
             f'{model}', 
             verticalalignment='bottom',
             horizontalalignment='left', fontsize=16)

# Adding labels and title
plt.xticks(bar_positions, strategies, fontsize=16)
plt.yticks(fontsize=16)
# plt.xlabel('Enhancement Strategies')
plt.ylabel('MAE [eV]', fontsize=18)
# plt.title('MAE for Different Enhancement Strategies with Error Bars')

# Show the plot
plt.tight_layout()
# plt.show()

# Save the plot
plt.savefig(f"bar_plots/barplot_mae.png", dpi=300, bbox_inches='tight')
```
