```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import numpy as np
# import matplotlib.pyplot as plt

# Categories and their corresponding MAE values
categories = ['Baseline', 'GAP', 'CA', 'GAP+CA']
mae_duplicate = [0.777, 0.671, 0.394, 0.352]
mae_unique = [0.555, 0.613, 0.401, 0.392]

# Setting the positions and width for the bars
pos = np.arange(len(categories))
bar_width = 0.35

# Creating the bar plot
# lighter_darkgreen = 'darkgreen' + '70'  # 70% opacity (0, 128, 0)
# lighter_maroon = 'maroon' + '70' #(128, 0, 0)
alpha_value=0.8
fig, ax = plt.subplots(figsize=(12, 8))
bar1 = ax.bar(pos - bar_width/2, mae_duplicate, 
              bar_width, label='Duplicate Texts', 
              alpha=alpha_value) #, color="darkgreen")
bar2 = ax.bar(pos + bar_width/2, mae_unique, 
              bar_width, label='Unique Texts', 
              alpha=alpha_value) #, color="maroon")

# Adding labels and title
ax.set_xlabel('')
ax.set_ylabel('MAE [eV]', fontsize=20)  # Increase fontsize of y-label
# ax.set_title('MAE Values for Duplicate and Unique Texts', fontsize=16)

# Display actual values at the bars
for bar in bar1 + bar2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',  # Format the value to 3 decimal places
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=16)

# Increase fontsize of x and y-axis ticks
ax.tick_params(axis='both', which='both', labelsize=18)
plt.xticks(pos, categories)
plt.legend(fontsize=18)

# Displaying the plot
plt.show()

fig.savefig(f"bar_plots/duplicates.png", dpi=300, bbox_inches='tight')
```

```python
# Data with reordered enhancements and corrected label
enhancements = ['Baseline', 'GAP', 'CA', 'GAP+CA']
categories = ['<s>', 'adsorbate', 'catalyst', 'configuration']
data = np.array([
    [0.045, 0.492, 0.137, 0.326],
    [0.012, 0.224, 0.257, 0.507],
    [0.034, 0.112, 0.344, 0.510],
    [0.017, 0.296, 0.157, 0.530]
])

# Setting the positions and width for the bars
pos = np.arange(len(enhancements))
bar_width = 0.2

# Creating the bar plot
fig, ax = plt.subplots(figsize=(10, 6))

for i, category in enumerate(categories):
    ax.bar(pos + i * bar_width, data[:, i], bar_width, label=category)

# Adding labels and title
ax.set_xlabel('Enhancement')
ax.set_ylabel('Values')
ax.set_title('Values for Different Categories across Enhancements')
ax.set_xticks(pos + (bar_width * (len(categories) - 1)) / 2)
ax.set_xticklabels(enhancements)
plt.legend()

# Displaying the plot
plt.show()
```
